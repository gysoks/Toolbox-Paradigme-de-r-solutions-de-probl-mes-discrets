Here is a simple example you find in the [test.py](./test.py) script. It runs an exhaustive serach of the convex hull for a randomly generated set of 8 points in a 2D space.

```python
from random import seed

from exhaustive import exhaustive
from utils import create_points, scatter_plot

import matplotlib.pyplot as pyplot

# initialize the random generator seed to always use the same set of points
seed(0)
# creates some points
pts = create_points(8)
show = False  # to display a frame
save = False  # to save into .png files in "figs" directory
scatter_plot(pts, [[]], title="convex hull : initial set", show=show, save=save)
print("Points:", pts)
# compute the hull
hull = exhaustive(pts, show=show, save=save)
print("Hull:", hull)
scatter_plot(pts, [hull], title="convex hull : final result", show=show, save=save)
```

Now, if you want to benchmark your algorithms, you can extend the [benchmark.py](https://gitlab.emse.fr/picard/PRPD/blob/master/convexhull/benchmark.py) script, to run your algorithms on randomly generated sets of points:

```python
from random import seed
from time import time

from exhaustive import exhaustive
from utils import *
from benchmark import benchmark

import matplotlib.pyplot as pyplot
        
seed(0)
algorithms = [exhaustive]

results = {}      

pyplot.xlabel('number of point')
pyplot.ylabel('average time (s)')

for algorithm in algorithms:
    results = benchmark(runs=10, sizes=range(2, 9, 2), method=algorithm)
    pyplot.plot(results['sizes'],results['avg time'],label=algorithm.__name__)
    
pyplot.legend()
pyplot.show()
```
